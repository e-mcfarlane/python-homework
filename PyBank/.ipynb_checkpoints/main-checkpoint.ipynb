```python
from pathlib import Path
import csv

filepath = Path("Resources/budget_data.csv")

pnldict = {}
header = ''

# transform CSV into a dictionary

with open(filepath,'r') as file:
    csvreader = csv.reader(file,delimiter=',')
    header = next(csvreader)
    for row in csvreader:
        pnldict[row[0]]=int(row[1])

# The total number of months included in the dataset

totMonths = 0

totMonths = len(pnldict)

# The net total amount of Profit/Losses over the entire period

plSum = 0

for v in pnldict.values():
    plSum = plSum + v

# Find the Greatest Increase and Decrease:

iKey = []

iValues = []

cList = []

cAvg = 0

highest = 0

lowest = 0

highKey = ''

lowKey = ''

for k,v in pnldict.items():
    iKey.append(k)
    iValues.append(v)

for i in range(1, totMonths):
    cList.append(iValues[i] - iValues[i-1])
    if cList[i-1] > highest:
        highest = cList[i-1]
        highKey = iKey[i]
    if cList[i-1] < lowest:
        lowest = cList[i-1]
        lowKey = iKey[i]

# The average of the changes in Profit/Losses over the entire period    
for i in cList:
    cAvg = cAvg + i

cAvg = (round(cAvg/(totMonths-1),2))

# Print to Terminal
print('Financial Analysis')
print('---------------------------------------')
print(f'Total Months: {totMonths}')
print(f'Average Change: ${cAvg}')
print(f'Greatest Increase in Profits: {highKey} (${highest})')
print(f'Greatest Decrease in Profits: {lowKey} (${lowest})')

# Output to File

output_path = Path("results.txt")
with open(output_path,"w") as f:
    f.write("Financial Analysis \n")
    f.write("--------------------------------------- \n")
    f.write(f"Total Months: {totMonths} \n")
    f.write(f"Average Change: ${cAvg} \n")
    f.write(f"Greatest Increase in Profits: {highKey} (${highest}) \n")
    f.write(f"Greatest Decrease in Profits: {lowKey} (${lowest})")
```
